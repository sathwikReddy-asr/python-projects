```python
df['age_group'] = pd.cut(df['victim_age'], [0, 20, 40, 60, 80],
                         labels=['0-20', '21-40', '41-60', '61-80'])
```

CONVERTING THE AGE_GROUP COLUMN INTO GROUPS

```python
age_group=pd.cut(df['victim_age'],[0,20,40,60,80])
table=df.pivot_table(values='id',index=['victim_gender',age_group],columns='crime_type')
percent_table = table.div(table.sum(axis=1), axis=0) * 100
print(percent_table)
```
